oporer 3d array ke chaile amra 2d tew convert korte pari,korar jonno sei array er modde koto ta element (6,3,4) ase setar jonno kono change jeno na hoi....mane chailew qmra (12,3) asob ba onno valu dite parbo na.

```python
lala=prb.reshape(12,6)
```

আমরা জানি মোট element = 72(6,3,4)
তাই reshape করার সময় নতুন shape-এর গুণফলও 72 হতে হবে।মানে প্রতি 6টা block-এর মধ্যে 3টা row, 4টা column ছিল।

এখন reshape করলে হবে:মানে:

6টা row থাকবে

প্রতি row-তে 12টা element থাকবে

অর্থাৎ আগের (3 × 4 = 12) অংশটা এক লাইনে পরিণত হচ্ছে। 

so ata aakta 2d array so 3d ke kibabe element thik rekhe 2d kora jai atai dekhano holo aseusome.
