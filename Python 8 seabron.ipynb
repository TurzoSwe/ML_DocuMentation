এই graph =
📌 Right-skewed distribution (positive skew)
📌 মানে বেশি data কম বয়সে, আর কিছু বড় বয়স tail বানিয়েছে

Normal distribution না কেন?

❌ Normal হলে:

একদম symmetric (bell shape) হতো

👉 কিন্তু এখানে:

ডান পাশে লম্বা tail → skewed


KDE line কী দেখাচ্ছে?

👉 যে smooth curve দেখছো (blue line):

এটা data-এর density
distribution shape বুঝতে help করে


বাম পাশে (low age) → data বেশি
ডান পাশে (high age) → লম্বা tail

👉 এই জন্য এটাকে বলে:
📌 right skewed

Tail = graph-এর সেই অংশ, যেখানে ডাটা কম থাকে কিন্তু লম্বা হয়ে ছড়িয়ে থাকে

📌 সহজভাবে:
👉 graph-এর “লেজ” অংশ

[10, 12, 15, 18, 20, 22, 25, 100]

👉 এখানে:

10–25 → main data
100 → অনেক দূরে

👉 এই “100” এর দিকটাই হলো tail

🔹 Normal Distribution কী?
👉 Normal Distribution (বা Gaussian distribution) হলো এমন একটা distribution যেখানে ডাটা মাঝখানে বেশি থাকে এবং দুই পাশে সমানভাবে কমে যায়।

দেখতে bell shape (ঘণ্টার মতো)
👉 একদম symmetric (দুই পাশ সমান)

Real-life example

👉 অনেক জিনিস normal distribution follow করে:

মানুষের height
exam marks
IQ score

JOIN PLOT

```python
sns.jointplot(data=data,x="total_bill",y="tip")
```

এই গ্রাফ একসাথে দুইটা জিনিস দেখায়:

relation (সম্পর্ক) → bill vs tip
distribution (বন্টন) → আলাদা আলাদা histogram

প্রতিটি ডট = একটি কাস্টমারের ডাটা
x-axis → total bill
y-axis → tip
👉 এখানে দেখা যাচ্ছে: বিল যত বেশি, টিপসও সাধারণত তত বেশি (positive relationship)   mazkhaner scatter plot er .

উপরের Histogram (total_bill)
শুধু total_bill এর distribution দেখাচ্ছে
কোন বিলের মান বেশি হয়েছে, সেটা বোঝা যায়

ডান পাশের Histogram (tip)
শুধু tip এর distribution দেখাচ্ছে
কোন টিপসের মান বেশি এসেছে তা বোঝা যায়

Scatter plot (স্ক্যাটার প্লট) হলো এমন একটি গ্রাফ যেখানে ডট (point) ব্যবহার করে দুইটা ভেরিয়েবলের সম্পর্ক দেখানো হয়।

By default ata scatter plot ea ase but atake amra aro bibinno plot ea convert koret pari.

Pairplot হলো Seaborn লাইব্রেরির একটি ফাংশন (sns.pairplot()), যেটা একসাথে অনেকগুলো ভেরিয়েবলের pairwise relationship (জোড়া জোড়া সম্পর্ক) দেখায়।

তোমার ডাটাতে আছে:


height


weight


age


👉 pairplot করলে কী হবে?


height vs weight


height vs age


weight vs age


➡️ সবগুলা combination একসাথে গ্রাফে দেখাবে!

🔹 গ্রাফটা কেমন হয়?
Pairplot আসলে একটা matrix (টেবিলের মতো):


📍 Diagonal (কোণাকুণি অংশ) → প্রতিটি ভেরিয়েবলের distribution (histogram / KDE)


📍 Off-diagonal (বাকি অংশ) → scatter plot (দুই ভেরিয়েবলের সম্পর্ক)

🔹 সংক্ষেপে
👉 Pairplot = অনেকগুলো scatter plot + histogram একসাথে


```python
import seaborn as sns
import matplotlib.pyplot as plt

# dataset load
df = sns.load_dataset("iris")

# pairplot তৈরি
sns.pairplot(df)

# show plot
plt.show()
```

এই কোড রান করলে:

সব numeric column (sepal_length, sepal_width, etc.) এর pairwise scatter plot দেখাবে
diagonal এ histogram দেখাবে
