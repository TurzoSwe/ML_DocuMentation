```python
ind_tuple=list(zip(groups, inside_groups))
ind_tuple
```

```python
index=pd.MultiIndex.from_tuples(tuples=ind_tuple)
#এটা ওই tuple গুলো দিয়ে 2-level index (hierarchical index) বানায়।
```

Level 1 → group1, group2
Level 2 → row1, row2, row3

কিন্তু pandas এ তাদের নাম নেই, তাই দেখাচ্ছে:

[None, None]
