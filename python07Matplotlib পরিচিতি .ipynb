plt.subplot(row, column, index)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 3, 5]

# 1st subplot
plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r--')
#--  disis karon red color er line ke dashed kore dibe
plt.title("Graph 1")

# 2nd subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b')
plt.title("Graph 2")

plt.show()
```

2 row মানে একটা উপরে + একটা নিচে

1 row (একটা সারি) মানে সব গ্রাফ এক লাইনে থাকবে (পাশাপাশি)

```python
plt.subplot(2, 1, 1)  # উপরের গ্রাফ
plt.plot(x, y1)

plt.subplot(2, 1, 2)  # নিচের গ্রাফ
plt.plot(x, y2)

plt.show()
```

Legend()method

legend() কী?

👉 legend() ব্যবহার করা হয়
গ্রাফে কোন লাইন কোনটা বোঝানোর জন্য (label দেখাতে)

🔹 Basic Idea

👉 আগে line-এ label দিতে হবে
👉 তারপর legend() call করতে হবে

```python
import matplotlib.pyplot as plt

x = [1,2,3]

plt.plot(x, [i**2 for i in x], label="Square")
plt.plot(x, [i**3 for i in x], label="Cube")

plt.legend()
plt.show()
```

👉 legend() = line-এর নাম দেখায় (label display করে)

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])#[left, bottom, width, height]
axes.plot(x,y,"black",alpha=0.5)  # alpha parameter is used to set the transparency of the line
plt.show()
```

HistoGram

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# গ্রাফকে সুন্দর (predefined style) দেখানোর জন্য এই স্টাইল ব্যবহার করা হয়েছে।

# make data
np.random.seed(1)#যাতে প্রতিবার একই র‍্যান্ডম ডাটা আসে friz kore dewa
x = 4 + np.random.normal(0, 1.5, 200)#২০০টা র‍্যান্ডম সংখ্যা তৈরি করছে
#👉 mean (গড়) = 0
#👉 standard deviation = 1.5
#4 + → সব ডাটাকে ৪ দিয়ে shift করা হয়েছে

#📌 ফলে ডাটা প্রায় ৪ এর আশেপাশে থাকবে।

# plot:
fig, ax = plt.subplots()#fig → পুরো গ্রাফের ক্যানভাস
#ax → যেখানে আমরা গ্রাফ আঁকব

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

np.random.seed(1) → যাতে প্রতিবার একই র‍্যান্ডম ডাটা আসে
np.random.normal(0, 1.5, 200) →
👉 ২০০টা র‍্যান্ডম সংখ্যা তৈরি করছে
👉 mean (গড়) = 0
👉 standard deviation = 1.5
4 + → সব ডাটাকে ৪ দিয়ে shift করা হয়েছে

📌 ফলে ডাটা প্রায় ৪ এর আশেপাশে থাকবে। fig → পুরো গ্রাফের ক্যানভাস ax → যেখানে আমরা গ্রাফ আঁকব

x → ডাটা
bins=8 → ডাটাকে ৮টা ভাগে ভাগ করবে
linewidth=0.5 → বারগুলোর বর্ডার পাতলা
edgecolor="white" → বর্ডারের রঙ সাদা

👉 Histogram মানে: ডাটা কতবার কোন রেঞ্জে পড়েছে সেটা দেখানো।


X-axis
xlim=(0, 8) → x-axis 0 থেকে 8
xticks=np.arange(1, 8) → 1 থেকে 7 পর্যন্ত টিক দেখাবে
Y-axis
ylim=(0, 56) → y-axis 0 থেকে 56
yticks=np.linspace(0, 56, 9) → 0 থেকে 56 পর্যন্ত ৯টা সমান ভাগ












Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

২৪টা random point তৈরি করা হচ্ছে
x এবং y দুইটাই random
mean ≈ 4 → তাই পয়েন্টগুলো ৪ এর আশেপাশে থাকবে

📌 মানে: 2D plane-এ random points


sizes → প্রতিটি ডটের size আলাদা
colors → প্রতিটি ডটের color আলাদা

📌 মানে: সব পয়েন্ট একই রকম না
x, y → পয়েন্টের position
s=sizes → প্রতিটি ডটের size
c=colors → color value
vmin, vmax → color scale range

Scatter plot =
📌 “ডাটার সম্পর্ক + pattern + outlier” একসাথে দেখার সবচেয়ে সহজ উপায়
